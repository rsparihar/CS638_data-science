```python
#Blocking plan :

#A, B --attr-equiv blocking--> candset(C0) --overlap-block--> candset(C1) |
#                                                                         | 
#A, B ------------rule-based-blocking--------------------> candset(C2)----|----union--->candset (C)
#                                                                         |
#A, B -----------black-box-blocking----------------------> candset(C3)----|
```

```python
# Attribute equivalence blocking: block A, B  using Runtime
ab = em.AttrEquivalenceBlocker()
C0 = ab.block_tables(A, B, 'Runtime', 'Runtime',
                    l_output_attrs=['Title', 'Runtime'], 
                    r_output_attrs=['Title', 'Runtime'])
C0
```

```python
# Overlap blocking over candidata set (C0)
ob = em.OverlapBlocker()

# block using name
C1 = ob.block_candset(C0, 'Title', 'Title', word_level=True, overlap_size=3)
```
